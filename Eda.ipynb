# EXPLORATORY DATA ANALYSIS - HR ATTRITION ANALYTICS

```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "Attrition cleaned data.csv"
df = pd.read_csv(file_path)

df.head()
```

# BASIC EDA

# Step 1: Understand Dataset Shape and Structure

```python
print("Dataset Shape:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
```

# Step 2: Univariate Analysis (One Feature at a Time)

Business Problem 1:
ðŸ”Ž What is the Age distribution of employees?

```python
plt.figure(figsize=(15,5))
# df["Age"].value_counts()
sns.histplot(data= df,x= df["Age"],kde= True,discrete= True,color= 'teal'
             )
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Business Problem 2:
ðŸ”Ž What is the gender-wise employee count?

```python
plt.figure(figsize=(8,5))
# df["Age"].value_counts()
sns.countplot(data= df,x= df["Gender"],palette="pastel")
plt.title('Gender Distribution')
plt.ylabel('frequency')
plt.show()
```

Business Problem 3:
ðŸ”Ž Attrition Rate (%)

```python
attrition_rate = df['Attrition'].value_counts(normalize=True) *100
print(attrition_rate)
```

# Step 3: Bivariate Analysis (Two Variables)

Business Problem 5:

# ADVANCED EDA

# Step 1: Root Cause Analysis for High Attrition Departments

 
Business Problem 1:
ðŸ”Ž Which department has highest Attrition % and what factors (Overtime, Job Satisfaction, Income) drive it?

```python
# Attrition Rate by Department
dept_attrition = pd.crosstab(df['Department'], df['Attrition'], normalize='index') * 100
print(dept_attrition)

# Further Analysis: Overtime within high-attrition departments
high_attr_dept = dept_attrition['Yes'].idxmax()
# high_attr_dept
subset = df[df['Department'] == high_attr_dept]
sns.countplot(x='OverTime', hue='Attrition', data=subset, palette='Set2')
plt.title(f'Attrition vs Overtime in {high_attr_dept} Department')
plt.show()
```

# Step 2: Multi-Dimensional Attrition Risk Segmentation


Business Problem 2:
ðŸ”Ž What is the combined impact of Age Group, Gender, and Job Role on Attrition?

```python
# Create Age Groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,18,25,35,45,60], labels=['0-18','18-25','26-35','36-45','46-60'])

pivot = pd.pivot_table(df, values='EmployeeID', index=['AgeGroup','Gender','JobRole'],
                       columns='Attrition', aggfunc='count', fill_value=0)

print(pivot.sort_values(by='Yes', ascending=False).head(30))
```

# Step 3: Promotion Stagnation Impact on Attrition


Business Problem 3:
ðŸ”Ž Are employees with more years since last promotion leaving more?

```python
sns.countplot(x=df["YearsSinceLastPromotion"], hue='Attrition', data=df, palette='Accent')
plt.title('Years Since Last Promotion vs Attrition')
plt.ylabel("frequency")
plt.show()
```

# Step 4: Work-Life Balance vs Overtime vs Attrition (Interaction Effect)
Business Problem 4:
ðŸ”Ž Is low work-life balance combined with overtime increasing attrition?

```python
plt.figure(figsize=(10,6))
sns.catplot(x='OverTime', hue='Attrition', col=df["WorkLifeBalance"], data=df, kind='count', palette='Set3')
plt.subplots_adjust(top=0.8)
plt.suptitle('Work Life Balance + Overtime Impact on Attrition')
plt.show()
```

# Step 5: Income Satisfaction Mismatch Analysis
Business Problem 5:
ðŸ”Ž Are high-income but low-satisfaction employees leaving?

```python
high_income = df['MonthlyIncome'] > df['MonthlyIncome'].quantile(0.75)
low_satisfaction = df['JobSatisfaction'] == "Bad"
df['Income_Satisfaction_Flag'] = high_income & low_satisfaction

sns.countplot(x='Income_Satisfaction_Flag', hue='Attrition', data=df, palette='Set1')
plt.title('Attrition for High Income + Low Satisfaction Employees')
plt.show()
```

# Step 6: Service Tenure Binning vs Attrition
Business Problem 6:
ðŸ”Ž Attrition across different service tenure bands

```python
df['TenureCategory'] = pd.cut(df['YearsAtCompany'],
                              bins=[0,1,3,5,10,20,40],
                              labels=['<1yr','1-3yrs','3-5yrs','5-10yrs','10-20yrs','20-40yrs'])
sns.countplot(x=df["TenureCategory"], hue='Attrition', data=df, palette='Accent')
plt.title('Attrition by Tenure Category')
plt.ylabel("frequency")
plt.show()
```
